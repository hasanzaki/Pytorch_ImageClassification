```
X = torch.Tensor([[1,0,1,0],[1,0,1,1],[0,1,0,1]])
y = torch.Tensor([[1],[1],[0]])
```

```
print(X)
```

```
print(y)
```

```
def sigmoid(x):
  return 1/(1+torch.exp(-x))
```

```
def derivatives_sigmoid(x):
  return sigmoid(x) * (1-sigmoid(x))
```

```
epoch=7000
lr = 0.1
inputSize = X.shape[1]
hiddenSize = 3
outputSize = 1
```

```
# randomly initialize weight and biases
wh = torch.randn(inputSize, hiddenSize).type(torch.FloatTensor)
bh = torch.randn(1, hiddenSize).type(torch.FloatTensor)
wout = torch.randn(hiddenSize,outputSize)
bout = torch.randn(1,outputSize)
```

```
for i in range(epoch):
  #forward
  hidden_layer_input1 = torch.mm(X, wh)
  hidden_layer_input = hidden_layer_input1 + bh
  hidden_layer_activations = sigmoid(hidden_layer_input)

  output_layer_input1 = torch.mm(hidden_layer_activations, wout)
  output_layer_input = output_layer_input1 + bout
  output = sigmoid(output_layer_input)

  # backprop
  E = y-output
  slope_output_layer = derivatives_sigmoid(output)
  slope_hidden_layer = derivatives_sigmoid(hidden_layer_activations)
  d_output = E * slope_output_layer
  Error_at_hidden_layer = torch.mm(d_output, wout.t())
  d_hiddenlayer = Error_at_hidden_layer * slope_hidden_layer

  # update weight & biases
  wout += torch.mm(hidden_layer_activations.t(), d_output)*lr
  bout += d_output.sum() * lr
  wh += torch.mm(X.t(), d_hiddenlayer) * lr
  bh += d_output.sum() * lr
```

```
print('actual:', y)
```

```
print('predicted:', output)
```
